# TextUtil
デュエルの情報に関するテキストを取得するクラス、TextUtilの使用例です。

```python
from ygo.util.text import TextUtil
from ygo.constants import EffectNo

# インスタンスを作成
text_util = TextUtil()   
```

## カードごとの関数呼び出し例

### 参照用のcard_id定義

```python
CID_BLUE_EYES       = 1004  # 「青眼の白龍」のcard_id
CID_HEAVY_STORM     = 1008  # 「大嵐」のcard_id
CID_MIRROR_FORCE    = 1013  # 「聖なるバリア －ミラーフォース－」のcard_id
CID_EXPLODER_DRAGON = 1018  # 「ボマー・ドラゴン」のcard_id
```

### カード名取得

```python
print(text_util.get_card_name(CID_BLUE_EYES))
print(text_util.get_card_name(CID_EXPLODER_DRAGON))
print(text_util.get_card_name(CID_HEAVY_STORM))
print(text_util.get_card_name(CID_MIRROR_FORCE))
```

### 読み仮名取得

```python
print(text_util.get_card_sort_name(CID_BLUE_EYES))
print(text_util.get_card_sort_name(CID_EXPLODER_DRAGON))
print(text_util.get_card_sort_name(CID_HEAVY_STORM))
print(text_util.get_card_sort_name(CID_MIRROR_FORCE))
```

### テキスト取得

```python
print(text_util.get_card_text(CID_BLUE_EYES))
print(text_util.get_card_text(CID_EXPLODER_DRAGON))
print(text_util.get_card_text(CID_HEAVY_STORM))
print(text_util.get_card_text(CID_MIRROR_FORCE))
```

### 効果テキスト取得

```python
print(text_util.get_effect_text(CID_EXPLODER_DRAGON, EffectNo.NUM2))
print(text_util.get_effect_text(CID_HEAVY_STORM, EffectNo.NUM1))
print(text_util.get_effect_text(CID_MIRROR_FORCE, EffectNo.NUM1))
```

### 属性取得

```python
print(text_util.get_card_attr(CID_BLUE_EYES))
print(text_util.get_card_attr(CID_EXPLODER_DRAGON))
print(text_util.get_card_attr(CID_HEAVY_STORM))
print(text_util.get_card_attr(CID_MIRROR_FORCE))
```

### レベル取得

```python
print(text_util.get_level(CID_BLUE_EYES))
print(text_util.get_level(CID_EXPLODER_DRAGON))
print(text_util.get_level(CID_HEAVY_STORM))
print(text_util.get_level(CID_MIRROR_FORCE))
```

### 種族取得

```python
print(text_util.get_species(CID_BLUE_EYES))
print(text_util.get_species(CID_EXPLODER_DRAGON))
print(text_util.get_species(CID_HEAVY_STORM))
print(text_util.get_species(CID_MIRROR_FORCE))
```

### 攻撃力取得

```python
print(text_util.get_atk(CID_BLUE_EYES))
print(text_util.get_atk(CID_EXPLODER_DRAGON))
print(text_util.get_atk(CID_HEAVY_STORM))
print(text_util.get_atk(CID_MIRROR_FORCE))
```

### 守備力取得

```python
print(text_util.get_def(CID_BLUE_EYES))
print(text_util.get_def(CID_EXPLODER_DRAGON))
print(text_util.get_def(CID_HEAVY_STORM))
print(text_util.get_def(CID_MIRROR_FORCE))
```

### 種類取得

```python
print(text_util.get_frame(CID_BLUE_EYES))
print(text_util.get_frame(CID_EXPLODER_DRAGON))
print(text_util.get_frame(CID_HEAVY_STORM))
print(text_util.get_frame(CID_MIRROR_FORCE))
```

### 魔法罠の種類取得

```python
print(text_util.get_icon(CID_BLUE_EYES))
print(text_util.get_icon(CID_EXPLODER_DRAGON))
print(text_util.get_icon(CID_HEAVY_STORM))
print(text_util.get_icon(CID_MIRROR_FORCE))
```

### ペンデュラムスケール取得（今回のプールでは値はありません）

```python
print(text_util.get_scale(CID_BLUE_EYES))
print(text_util.get_scale(CID_EXPLODER_DRAGON))
print(text_util.get_scale(CID_HEAVY_STORM))
print(text_util.get_scale(CID_MIRROR_FORCE))
```
