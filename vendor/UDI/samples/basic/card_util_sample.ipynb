# CardUtil
カードの情報を取得するクラス、CardUtilの使用例です。

```python
from ygo.util.card import CardUtil
from ygo.constants import EffectNo

# インスタンスを作成（中でdataファイルを読み込んでいます）
card_util = CardUtil()   
```

## 関数呼び出し例

### 参照用のcard_id定義

```python
CID_BLUE_EYES       = 1004  # 「青眼の白龍」のcard_id
CID_HEAVY_STORM     = 1008  # 「大嵐」のcard_id
CID_MIRROR_FORCE    = 1013  # 「聖なるバリア －ミラーフォース－」のcard_id
CID_EXPLODER_DRAGON = 1018  # 「ボマー・ドラゴン」のcard_id
```

### カード名取得

```python
print(card_util.get_name(CID_BLUE_EYES))
print(card_util.get_name(CID_EXPLODER_DRAGON))
print(card_util.get_name(CID_HEAVY_STORM))
print(card_util.get_name(CID_MIRROR_FORCE))
```

### 読み仮名取得

```python
print(card_util.get_sort_name(CID_BLUE_EYES))
print(card_util.get_sort_name(CID_EXPLODER_DRAGON))
print(card_util.get_sort_name(CID_HEAVY_STORM))
print(card_util.get_sort_name(CID_MIRROR_FORCE))
```

### テキスト取得

```python
print(card_util.get_text(CID_BLUE_EYES))
print(card_util.get_text(CID_EXPLODER_DRAGON))
print(card_util.get_text(CID_HEAVY_STORM))
print(card_util.get_text(CID_MIRROR_FORCE))
```

### 効果テキスト取得

```python
print(card_util.get_effect_text(CID_EXPLODER_DRAGON, EffectNo.NUM2))
print(card_util.get_effect_text(CID_HEAVY_STORM, EffectNo.NUM1))
print(card_util.get_effect_text(CID_MIRROR_FORCE, EffectNo.NUM1))
```

### 属性取得
enum（`constants.Attribute`）で取得しています

```python
print(card_util.get_attribute(CID_BLUE_EYES).name)
print(card_util.get_attribute(CID_EXPLODER_DRAGON).name)
print(card_util.get_attribute(CID_HEAVY_STORM).name)
print(card_util.get_attribute(CID_MIRROR_FORCE).name)
```

### レベル取得

```python
print(card_util.get_level(CID_BLUE_EYES))
print(card_util.get_level(CID_EXPLODER_DRAGON))
print(card_util.get_level(CID_HEAVY_STORM))
print(card_util.get_level(CID_MIRROR_FORCE))
```

### 種族取得
enum（`constants.Species`）で取得しています

```python
# 種族取得
print(card_util.get_species(CID_BLUE_EYES).name)
print(card_util.get_species(CID_EXPLODER_DRAGON).name)
print(card_util.get_species(CID_HEAVY_STORM).name)
print(card_util.get_species(CID_MIRROR_FORCE).name)
```

### 攻撃力取得

```python
print(card_util.get_atk(CID_BLUE_EYES))
print(card_util.get_atk(CID_EXPLODER_DRAGON))
print(card_util.get_atk(CID_HEAVY_STORM))
print(card_util.get_atk(CID_MIRROR_FORCE))
```

### 守備力取得

```python
print(card_util.get_def(CID_BLUE_EYES))
print(card_util.get_def(CID_EXPLODER_DRAGON))
print(card_util.get_def(CID_HEAVY_STORM))
print(card_util.get_def(CID_MIRROR_FORCE))
```

### 種類取得

```python
print(card_util.get_frame(CID_BLUE_EYES))
print(card_util.get_frame(CID_EXPLODER_DRAGON))
print(card_util.get_frame(CID_HEAVY_STORM))
print(card_util.get_frame(CID_MIRROR_FORCE))
```

### 魔法・罠かどうかを取得

```python
print(card_util.is_magic_trap(CID_BLUE_EYES))
print(card_util.is_magic_trap(CID_EXPLODER_DRAGON))
print(card_util.is_magic_trap(CID_HEAVY_STORM))
print(card_util.is_magic_trap(CID_MIRROR_FORCE))
```

### 魔法罠の種類取得

```python

print(card_util.get_icon(CID_BLUE_EYES))
print(card_util.get_icon(CID_EXPLODER_DRAGON))
print(card_util.get_icon(CID_HEAVY_STORM))
print(card_util.get_icon(CID_MIRROR_FORCE))
```

### ペンデュラムスケール取得

```python
print(card_util.get_scale(CID_BLUE_EYES))
print(card_util.get_scale(CID_EXPLODER_DRAGON))
print(card_util.get_scale(CID_HEAVY_STORM))
print(card_util.get_scale(CID_MIRROR_FORCE))
```

### ペンデュラムモンスターかどうかを取得

```python

print(card_util.is_pendulum(CID_BLUE_EYES))
print(card_util.is_pendulum(CID_EXPLODER_DRAGON))
print(card_util.is_pendulum(CID_HEAVY_STORM))
print(card_util.is_pendulum(CID_MIRROR_FORCE))
```
